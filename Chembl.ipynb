Функция для получения смайлс, где все фторы заменены на водород
И функция, считающая количество фторов

```python
def get_H(x) -> str:
    a = str.replace(x, 'F', '')
    return str.replace(a, '()', '')
def count_F(x) -> int:
    return x.count('F')
df['Smiles_H'] = df['Smiles'].apply(lambda x : get_H(x))
df['count_F'] = df['Smiles'].apply(lambda x : count_F(x))
```

Функция создает три массива словарей (в целом, это костыль, но он работает со сложностью С x О(n), поэтому не страшно): для ID водородсодержащих соединений, для их молярной массы и индекса из таблицы, которому это соединение соответствует

Поиск по Smiles находит кучу соединений, которые нам не подходят. Для их отсеивания мы смотрим, сколько целых раз заменился фтор на водород. Причем для хорошего совпадения (ведь молярная масса водорода чуть больше целого (1.008), а фтора -- чуть меньше (18.99)) необходимо убедиться, что вычитание из дельты единицы и прибавления ее же приводит нас к положительному результату

```python
ID_H = []
MWT_H = []
index = []
def ID_MWT_H(x, y, z, ID_H = ID_H, MWT_H = MWT_H, index = index):
    try:
        substructure = new_client.substructure
        res = substructure.filter(
        smiles=x).filter(
        molecule_properties__mw_freebase__lte=y).only(
        ['molecule_chembl_id','molecule_properties'])
        for i in range(len(res)):
            delta= y - float(res[i].get('molecule_properties').get('full_mwt'))
            if ((delta +1) // 18 == float(z) and (delta - 1) // 18 + 1 == float(z)):
        
                index.append(x)
                ID_H.append(res[i].get('molecule_chembl_id'))
                MWT_H.append(res[i].get('molecule_properties').get('full_mwt'))
                
                
                return res[i].get('molecule_chembl_id')
                break
    except:
        print('Not found')
```
