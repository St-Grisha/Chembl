Функция для получения смайлс, где все фторы заменены на водород
И функция, считающая количество фторов

```python
def get_H(x) -> str:
    a = str.replace(x, 'F', '')
    return str.replace(a, '()', '')
def count_F(x) -> int:
    return x.count('F')
df_F['Smiles_H'] = df_F['Smiles'].apply(lambda x : get_H(x))
df_F['count_F'] = df_F['Smiles'].apply(lambda x : count_F(x))
```

Переименование колонок для возможности обращения к ним

```python
df_F = df_F.rename(columns={'Molecular Weight (Monoisotopic)' : 'MWT_F', 'ChEMBL ID' : 'ID_F', 'Molecular Formula' : 'Brutto_F'})
df = df.rename(columns={'Molecular Weight (Monoisotopic)' : 'MWT', 'ChEMBL ID' : 'ID', 'Molecular Formula' : 'Brutto'})
```

```python
a = pd.merge(df_F, df, how = 'inner', left_on = 'Smiles_H', right_on='Smiles')
```

```python
a.columns
```

```python
res = a.drop(['Bioactivities_x', 'MWT_F','Brutto_F','Smiles_H', 'count_F','Bioactivities_y','MWT',
       'Brutto'], axis = 1)
```

```python
res = res.rename({'ID_F' : 'ID_F', 'Smiles_x' : 'Smiles_F', 'ID' : 'ID_H', 'Smiles_y' : 'Smiles_H'}, axis=1)
```
