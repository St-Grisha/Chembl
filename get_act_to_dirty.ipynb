# Поиск активности на первых ста результатах

```python
df = pd.read_csv("ID.csv")
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
df
```

```python
from chembl_webresource_client.new_client import new_client
```

```python
act = pd.DataFrame()
def activity_F(x, act = act, activity = new_client.activity) -> pd.DataFrame:
    res = activity.filter(molecule_chembl_id=x).only(['assay_chembl_id', 'type',
                                                                'value'])
    if len(res) == 0:
        print('passive')
        return None
    else:
        res = pd.DataFrame(res)
        res['ID_F'] = x 
        act = pd.concat([act, res])
        print('active')
        return pd.DataFrame(act)
```

```python
df_train = df[200:300]
```

```python
%%time
F = df['ID_F'].apply(lambda x: activity_F(x))
```

```python
#F
```

```python
activ = F[0]
for i in range(1, len(F)):
    activ = pd.concat([activ, F[i]])
activ.to_csv('active66300.csv')
df_act = pd.merge(df, activ, on = 'ID_F', how = 'left')
df_act = df_act.rename(columns={'type' : 'type_F', 'value' : 'value_F'})
df_act = df_act.drop(df_act[df_act["assay_chembl_id"].isna()].index)
df_act
```

```python
df_act[['ID_F', 'assay_chembl_id', 'type_F', 'value_F']].to_csv('activity_F_66300.csv')
```

```python
def activity_H(x, act = act, activity = new_client.activity):
    res = activity.filter(molecule_chembl_id=x).only(['assay_chembl_id', 'type',
                                                                'value'])
    if len(res) == 0:
        #print('passive')
        return None
    else:
        res = pd.DataFrame(res)
        res['ID_H'] = x
        #print('active')
        return pd.DataFrame(res)
```

```python
%%time
activ_H = activity_H(df_act['ID_H'].unique()[0])
count = 0
for i in (df_act['ID_H'].unique()[1:]):
    activ_H = pd.concat([activ_H, activity_H(i)])
    count += 1
    if count % 100 == 0:
        print(count)
```

```python
activ_H.to_csv('activ_H.csv')
```

```python
activ_H = activ_H.rename(columns={'type' : 'type_H', 'value' : 'value_H'})
activ_H.to_csv('activ_H.csv')
```

```python
IC50_H = activ_H[activ_H['type_H'] == 'IC50']
IC50_H
```

```python
IC50_F = df_act[df_act['type_F'] == 'IC50']
```

```python
total = pd.merge(IC50_F, IC50_H, on = 'assay_chembl_id', how = 'left')
total = total[total['ID_H_x'] == total['ID_H_y']]
total = total.drop(['ID_H_y'], axis = 1)
```

```python
total = total.drop(['Smiles_H', 'Smiles_F', 'type_F'], axis = 1)
total = total.dropna()
total
```
